```python
import pandas as pd
```

```python
movies = pd.read_csv(
    "https://data-bootcamp-ztc.s3.amazonaws.com/movies_complete_cleaned.csv"
)

awards = pd.read_csv(
    "https://data-bootcamp-ztc.s3.amazonaws.com/parsed_awards_table.csv"
)

features = pd.read_csv(
    "../../resources/cleaned_data/plot_features.csv"
)
movies_sel = movies[["name", "genre_kaggle"]]

awards_sel = awards[["name", "oscar wins and nominations"]]

features_sel = features[['name', 'plot', 'idf_token']]


df = pd.merge(pd.merge(movies_sel, awards_sel, how="right", on="name"), features_sel, how = "right", on ="name")

df["target"] = [1 if x != 0 else 0 for x in df["oscar wins and nominations"]] #binary where movie was oscar nominated

df = df[["name", "target","plot","idf_token","genre_kaggle"]]

df.set_index("name", inplace = True)

genre_encoded = pd.get_dummies(df.genre_kaggle, prefix = "genre")

df = pd.merge(df, genre_encoded, how = "right", on = "name")


df.head()
```

```python
# X = 

y = df.target.values.reshape(-1,1)
```

```python
X = df.drop(["target","plot","genre_kaggle"],axis = 1)
X

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
